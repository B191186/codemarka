```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5,2)
plt.subplot(3,1,1)
plt.plot(x,x*2,'r-.')
plt.plot(x,x*3,'g--')
plt.plot(x,x*4,'y--')
plt.legend(["x", "y", "z"])
plt.show()
plt.subplot(3,1,2)
plt.plot(x,x*2,'r--')
plt.plot(x,x*3,'g--')
plt.plot(x,x*4,'y--')
plt.legend(["x", "y", "z"])
plt.show()
plt.subplot(3,1,3)
x=np.arange(0,5,0.2)
plt.plot(x,x*2,'r--',x,x*3,'gs',x,x*4,'y^')
plt.show()

```
